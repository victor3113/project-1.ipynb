```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
data = pd.read_csv('../taller/user_loggedin_teacher1.csv', delimiter=',')
```

```python
data.info()
```

```python
data['department'].unique()
```

# Punto 2

```python
# Calcular el total de logins por departamento
total_logins_por_department = data['department'].value_counts()
total_logins_por_department
```

```python
# Graficar los resultados
plt.figure(figsize=(12, 6))
total_logins_por_department.plot(kind='bar')
plt.title('Total Logins by Department')
plt.xlabel('Department')
plt.ylabel('Total Logins')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar el gráfico
plt.tight_layout()
plt.show()

# Imprimir el total de logins por departamento
print(total_logins_por_department)
```

# Punto 3

```python
#agrupamos los datos del CSV según los campos id (que identifica a cada docente) 
# y department (que indica el programa o departamento al que pertenece)y lo agrupamos con la funcion groupby.
login_counts_by_teacher = data.groupby(['id', 'department']).size().reset_index(name='total_logins')
print(login_counts_by_teacher)
```

```python
# Guardar los resultados en un archivo CSV
login_counts_by_teacher.to_csv('login_counts_by_teacher.csv', index=False)
```

# Punto 5

```python
# Filtrar departamentos no nulos y convertir a minúsculas para uniformidad
data = data.dropna(subset=['department'])
data['department'] = data['department'].str.lower()

# Calcular el número de logins por departamento
logins_por_departamento = data.groupby('department').size().reset_index(name='total_logins')

# Graficar el boxplot
plt.figure(figsize=(12, 8))
sns.boxplot(x='total_logins', y='department', data=logins_por_departamento, palette='viridis')
plt.title('Distribución de Inicios de Sesión por Departamento')
plt.xlabel('Total de Logins')
plt.ylabel('Departamento')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Punto 6

```python
# Agrupa por departamento y calcula el promedio de inicios de sesión
login_avg_by_department = data.groupby('department')['timecreated'].mean()

# Muestra el promedio de inicio de sesión por departamento
print(login_avg_by_department)
```

# Punto 7

```python
# Incluir solo registros donde el rol es "editingteacher"
teachers_data = data[data['rol'] == 'editingteacher']

# Calcular los cuartiles para la columna 'timecreated' agrupados por 'department'
quartiles_by_department = teachers_data.groupby('department')['timecreated'].quantile([0.25, 0.5, 0.75]).unstack()

# Mostrar los cuartiles por departamento
print(quartiles_by_department)
```
